```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image, ImageOps
import os
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage import filters
from skimage.restoration import denoise_bilateral, wiener
import cv2
from scipy.ndimage import gaussian_filter
from scipy.signal import unit_impulse
from scipy.fft import fft2, ifft2, fftshift, ifftshift
```

```python
types = ['sp_noise', 'speckle_noise', 'gaussian_noise', 'gaussian_blur', 'motion_blur', 'input_imgs']
img_names = os.listdir('stage1_data/input_imgs')
```

```python
# Loads single noise, clean img pair
def load_data(noise_type, img_name, method):
    noise_path = 'stage1_data/'+noise_type+'/'+img_name
    clean_path = 'stage1_data/input_imgs/'+img_name
    
    if method == 'pil':
        X_noise = np.array(Image.open(noise_path))/255
        y_true = np.array(Image.open(clean_path))/255
    elif method == 'cv2':
        X_noise = cv2.imread(noise_path)
        y_true = cv2.imread(clean_path)
    
    return X_noise, y_true
```

# Salt and Pepper Noise
Goals (PSNR/SSIM):
- 26.5/0.90

```python
# Salt Pepper Noise
k_size = 3
psnr_all = []
ssim_all = []
noise_type = 'sp_noise'

for img_name in img_names:
    X_noise, y_true = load_data(noise_type, img_name, 'pil')

    # Try simple median filter
    y_filt = []
    for c in range(3):
        y_filt.append(filters.median(X_noise[:,:,c], footprint=np.ones((k_size,k_size))))
    
    y_filt = np.stack(y_filt, axis=-1)
    psnr_all.append(psnr(y_true, y_filt))
    ssim_all.append(ssim(y_true, y_filt, data_range=1, channel_axis=2))

print('************** Salt and Pepper Noise **************')
print('Average PSNR: ' + str(np.mean(psnr_all)))
print('Average SSIM: ' + str(np.mean(ssim_all)))
```

```python
plt.imshow(X_noise)
plt.title('Noise Image')
plt.show()
# plt.imshow(y_true)
# plt.title('True Image')
# plt.show()
plt.imshow(y_filt)
plt.title('Filtered Image')
plt.show()
```

# Gaussian Blur
Goals (PSNR/SSIM):
- 20.5/0.65

```python
# Gaussian Blur
psnr_all = []
ssim_all = []
noise_type = 'gaussian_blur'

# Find Best Sigma for some Kernel Size
k_size = 13
X_noise, y_true = load_data(noise_type, img_names[0], 'pil')
for sigma in np.arange(0,k_size,0.1):
    psf = gaussian_filter(unit_impulse((k_size, k_size),idx='mid'), sigma=sigma)
    y_filt = []
    for c in range(3):
        y_filt.append(wiener(X_noise[:,:,c], psf, 0.1))
    
    y_filt = np.stack(y_filt, axis=-1)
    psnr_all.append(np.mean(psnr(y_true, y_filt)))
    ssim_all.append(np.mean(ssim(y_true, y_filt, data_range=1, channel_axis=-1)))

sigma = np.arange(0,k_size,0.1)[np.argmax(np.linalg.norm(np.array([psnr_all, ssim_all]),axis=0))]
psf = gaussian_filter(unit_impulse((k_size, k_size),idx='mid'), sigma=sigma)
psnr_all = []
ssim_all = []
# Wiener filter with finalized Sigma
for img_name in img_names:
    X_noise, y_true = load_data(noise_type, img_name, 'pil')

    y_filt = []
    for c in range(3):
        y_filt.append(wiener(X_noise[:,:,c], psf, 0.1))
    
    y_filt = np.stack(y_filt, axis=-1)
    y_filt[y_filt < 0.0] = 0.0
    y_filt[y_filt > 1.0] = 1.0
    psnr_all.append(psnr(y_true, y_filt))
    ssim_all.append(ssim(y_true, y_filt, data_range=1, channel_axis=-1))

print('************** Gaussian Blur **************')
print('Average PSNR: ' + str(np.mean(psnr_all)))
print('Average SSIM: ' + str(np.mean(ssim_all)))
```

```python
plt.imshow(X_noise)
plt.title('Noise Image')
plt.show()
# plt.imshow(y_true)
# plt.title('True Image')
# plt.show()
plt.imshow(y_filt)
plt.title('Filtered Image')
plt.show()
```

# Motion Blur
Goals (PSNR/SSIM):
- 21.5/0.70

```python
# Motion Blur
from skimage.restoration import wiener

noise_type = 'motion_blur'
psnr_all = []
ssim_all = []
y_all = []
kernel_size = 19
# Try wiener filter
for img_name in tqdm(img_names):
    X_noise, y_true = load_data(noise_type, img_name, 'pil')
    psnr_img, ssim_img = [],[]
    y_img = []
    
    for angle in np.arange(0,180,5):
        motion_blur_kernel = np.zeros((kernel_size, kernel_size))
        motion_blur_kernel[int((kernel_size-1)/2), :] = np.ones(kernel_size)
        rotation_matrix = cv2.getRotationMatrix2D((int((kernel_size-1)/2), int((kernel_size-1)/2)), angle, 1)
        motion_blur_kernel = cv2.warpAffine(motion_blur_kernel, rotation_matrix, (kernel_size, kernel_size))/kernel_size
        y_filt = []
        for c in range(3):
            h = wiener(X_noise[:,:,c], motion_blur_kernel, 0.78)
            y_filt.append(unsharp_mask(h, radius=1, amount=2))

        y_filt = np.stack(y_filt, axis=-1)
        y_filt[y_filt < 0.0] = 0.0
        y_filt[y_filt > 1.0] = 1.0

        psnr_img.append(np.mean(psnr(y_true, y_filt)))
        ssim_img.append(np.mean(ssim(y_true, y_filt, data_range=1, channel_axis=-1)))
        y_img.append(y_filt)

    idx = np.argmax(np.linalg.norm(np.array([psnr_img, ssim_img]),axis=0))
    
    psnr_all.append(psnr_img[idx])
    ssim_all.append(ssim_img[idx])
    y_all.append(y_img[idx])
    
print('************** Motion Blur **************')
print('Average PSNR: ' + str(np.mean(psnr_all)))
print('Average SSIM: ' + str(np.mean(ssim_all)))
```

### Motion Blur: Striations
Striations can occur when filtering with a large kernel and setting correlation boundary to 'fill'. Setting to 'wrap' results in no striations.

```python
noise_type = 'motion_blur'
X_noise, y_true = load_data(noise_type, img_names[0], 'pil')
kernel_size = 23
angle = 45
motion_blur_kernel = np.zeros((kernel_size, kernel_size))
motion_blur_kernel[int((kernel_size-1)/2), :] = np.ones(kernel_size)
rotation_matrix = cv2.getRotationMatrix2D((int((kernel_size-1)/2), int((kernel_size-1)/2)), angle, 1)
motion_blur_kernel = cv2.warpAffine(motion_blur_kernel, rotation_matrix, (kernel_size, kernel_size))
motion_blur_kernel /= np.sum(motion_blur_kernel)

from scipy.signal import correlate2d
noisy = []
for i in range(3):
    #noisy.append(correlate2d(y_true[:,:,i], motion_blur_kernel, mode='same', boundary='fill'))
    noisy.append(correlate2d(y_true[:,:,i], motion_blur_kernel, mode='same', boundary='fill'))
    
noisy = np.stack(noisy, axis=-1)

ssim_all = []
y_filt = []
snr = 50
for c in range(3):
    y_filt.append(wiener(noisy[:,:,c], motion_blur_kernel, 0.5))
    #y_filt.append(wiener_custom(noisy[:,:,c], motion_blur_kernel, snr))

y_filt = np.stack(y_filt, axis=-1)
y_filt[y_filt<0.0] = 0.0
y_filt[y_filt>1.0] = 1.0

plt.imshow(y_filt)

# np.mean(psnr(y_true, y_filt))
np.mean(ssim(y_true, y_filt, data_range=1, channel_axis=-1))
```

### Motion Blur: Striations Low Pass Filtering
Try removal of striations by low pass filter. Blurs image too much to be viable

```python
def low_pass(y_filt):
    y_out = []
    for c in range(3):
        F = fftshift(fft2(y_filt[:,:,c]))
        offset_0 = 20
        center = int(F.shape[0]/2)
        F[(center+offset_0):,:] = 0
        F[:(center-offset_0),:] = 0
        
        offset_1 = 80
        center = int(F.shape[1]/2)
        F[:,(center+offset_1):] = 0
        F[:,:(center-offset_1)] = 0
        
        y_out.append(np.real(ifft2(ifftshift(F))))
        
    return np.stack(y_out, axis=-1)

plt.plot( np.abs(fftshift(fft2(y_filt[:,:,0])))[:,696] )
plt.title('Vertical Cross-Section of FFT(y_filt)')
plt.show()
plt.plot( np.abs(fftshift(fft2(y_filt[:,:,0])))[256,:] )
plt.title('Horizontal Cross-Section of FFT(y_filt)')
plt.show()


y_out = low_pass(y_filt)
plt.imshow(y_out)
plt.show()
```

### Motion Blur: Delta Function
When applying motion blur to the delta function, fails to recover it given the larger motion blur kernel

```python
# Experiment with Weiner
kernel_size = 100
psf = unit_impulse((kernel_size, kernel_size),idx='mid')
#psf = np.zeros((kernel_size, kernel_size))
#psf[int((kernel_size-1)/2), :] = np.ones(kernel_size)

kernel_size = 23
angle = 135
motion_blur_kernel = np.zeros((kernel_size, kernel_size))
motion_blur_kernel[int((kernel_size-1)/2), :] = np.ones(kernel_size)
rotation_matrix = cv2.getRotationMatrix2D((int((kernel_size-1)/2), int((kernel_size-1)/2)), angle, 1)
motion_blur_kernel = cv2.warpAffine(motion_blur_kernel, rotation_matrix, (kernel_size, kernel_size))
motion_blur_kernel /= np.sum(motion_blur_kernel)

from scipy.signal import correlate2d
from scipy.fft import ifftshift
noisy = correlate2d(psf, motion_blur_kernel, mode='same', boundary = 'fill')

# Wiener by hand
snr = 100
Y = fft2(noisy)
H = fft2(motion_blur_kernel, s=noisy.shape)
filt=Y/H * np.abs(H)**2 / (np.abs(H)**2 + 1/snr)

filt = np.real(ifft2(filt))

# Wiener Skimage
filt = wiener(noisy, motion_blur_kernel, 0.01)

filt /= np.max(filt)
filt[filt<0.0] = 0.0
plt.imshow(filt)
```
